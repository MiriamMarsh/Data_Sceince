```python
features = ['Healthy life expectancy','Freedom to make life choices','Social support','GDP per capita']
```

```python
X = df[features]
y = df['Score']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_test,y_test)
```

```python
predictions = lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```

```python
lr2=LinearRegression()
```

```python
lr2.fit(X_train, y_train)
```

```python
predictions = lr2.predict(X_test)
```

```python
lr2.score(X_test, y_test)
```

```python
plt.scatter(predictions, predictions - y_test, c= 'g', s= 40)
plt.hlines(y=0, xmin = 0, xmax = 20000) 
plt.title('Residual plot')
plt.ylabel('Residual')
```

Our model is a good model since there is not such a big difference between the target and the prediction. The difference is only between -1 and 1. The accuracy score is 77.8%. I don't see to much of a pattern which shows a good graph. 

```python
MSE = mean_squared_error(y_test, predictions)
print(MSE)
```
